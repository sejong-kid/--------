```python
from konlpy.tag import Komoran,Okt, Hannanum
import re

komoran = Komoran()
okt = Okt()
han = Hannanum()

data = pd.read_excel('data14000_코모란.xlsx')
titles= data['title']
contents = data['content']

okl_con = []
kol_con = []
han_con = []
for c,i in enumerate(contents):
    try:
        i=re.sub('[^ㄱ-힣 ]',' ',str(i))
        okl_con.append([x for x in okt.nouns(str(i)) if len(x)>1])
        kol_con.append([x for x in komoran.nouns(str(i)) if len(x)>1])
        han_con.append([x for x in han.nouns(str(i)) if len(x)>1])
    except:
        pass
    print(c/len(contents))
    
okl_con_flatten = []
kol_con_flatten = []
han_con_flatten = []
all_flatten_con = []
all_l_con = okl_con + kol_con + han_con 
for i in all_l_con:
    all_flatten_con+=i
for x in okl_con:
    okl_con_flatten+=x
for x in kol_con:
    kol_con_flatten+=x
for x in han_con:
    han_con_flatten+=x
sorted_all_con = sorted(Counter(all_flatten_con).items(), key = lambda x : -x[1])
okt_con_gt4 = [x for x,y in Counter(okl_con_flatten).items() if len(x)>3]
kol_con_gt4 = [x for x,y in Counter(kol_con_flatten).items() if len(x)>3]
han_con_gt4 = [x for x,y in Counter(han_con_flatten).items() if len(x)>3]

okt_con_gt4_scattered = []
for i in okt_con_gt4:
    ko = komoran.nouns(i)
    h = han.nouns(i)
    if len(ko)!=1:
        okt_con_gt4_scattered+=ko
    if len(h)!=1:
        okt_con_gt4_scattered+=h
okt_con_gt4_scattered = [x for x in okt_con_gt4_scattered if len(x)>1]    

kol_con_gt4_scattered = []
for i in kol_con_gt4:
    ot = okt.nouns(i)
    h = han.nouns(i)
    if len(ot)!=1:
        kol_con_gt4_scattered+=ot
    if len(h)!=1:
        kol_con_gt4_scattered+=h
kol_con_gt4_scattered = [x for x in kol_con_gt4_scattered if len(x)>1]  

han_con_gt4_scattered = []
for i in han_con_gt4:
    ot = okt.nouns(i)
    ko = komoran.nouns(i)
    if len(ot)!=1:
        han_con_gt4_scattered+=ot
    if len(ko)!=1:
        han_con_gt4_scattered+=ko
han_con_gt4_scattered = [x for x in han_con_gt4_scattered if len(x)>1]   
print(len(han_con_gt4),len(kol_con_gt4), len(okt_con_gt4))

vocab = Counter(han_con_gt4_scattered+kol_con_gt4_scattered+okt_con_gt4_scattered+okl_con_flatten+kol_con_flatten+han_con_flatten).keys()
##vocab에서 길이 2-6까지의 단어를 토큰으로 인정
##코모란이 추출한 띄워쓰기가 있는 단어는 okt와 한나눔이 처리했으므로 띄워쓰기 있는 단어는 제거
new_vocab = []
for i in vocab:
    if ' ' not in i:
        if len(i)<7:
            new_vocab.append(i)
print(len(new_vocab))
```
